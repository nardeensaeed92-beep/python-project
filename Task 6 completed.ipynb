```python
# Part 1: Basic NumPy Operations
# Part 2: Basic Pandas Operations

import numpy as np
import pandas as pd

# ----------------------------
# Part 1: Basic NumPy Operations
# ----------------------------

# Create arrays
A = np.array([10, 20, 30, 40, 50])
B = np.array([5, 4, 3, 2, 1])

print("Array A:", A)
print("Array B:", B)

# Perform Operations
print("\n--- NumPy Basic Operations ---")
print("A + B =", A + B)
print("A - B =", A - B)
print("A * B =", A * B)      # element-wise multiplication
print("A / B =", A / B)      # element-wise division

# Apply NumPy Functions
print("\n--- NumPy Functions ---")
print("Mean of A:", np.mean(A))
print("Max of A:", np.max(A))
print("Min of A:", np.min(A))
print("Dot Product of A and B:", np.dot(A, B))

# Reshape A into 5x1 matrix
A_reshaped = A.reshape(5, 1)
print("A reshaped into 5x1 matrix:\n", A_reshaped)

# ----------------------------
# Part 2: Basic Operations with Pandas
# ----------------------------

students_data = {
    "Name": ["Ali", "Sara", "Omar", "Laila", "Khaled"],
    "Age": [20, 21, 19, 22, 20],
    "Grade": ["A", "B", "A", "C", "B"],
    "Score": [88, 76, 95, 67, 80]
}

df = pd.DataFrame(students_data)

print("\n--- Pandas DataFrame ---")
print(df)

# Accessing Data
print("\n--- Accessing Data ---")
print("Names column:\n", df["Name"])
print("First 3 rows:\n", df.head(3))
print("Row with index 2:\n", df.loc[2])

# Basic Operations
print("\n--- Pandas Basic Operations ---")
print("Average Score:", df["Score"].mean())
print("Maximum Score:", df["Score"].max())
print("Minimum Score:", df["Score"].min())

# Filter students with Score > 80
print("\nStudents with Score > 80:")
print(df[df["Score"] > 80])
```
