```python
# library_system.py

from typing import List

class Book:
    def _init_(self, title: str, author: str, isbn: str, available: bool = True):
        self.title = title
        self.author = author
        self.__isbn = isbn            # private attribute
        self.available = available

    # getter for isbn
    def get_isbn(self) -> str:
        return self.__isbn

    # setter for isbn (if ever needed)
    def set_isbn(self, new_isbn: str):
        self.__isbn = new_isbn

    def display_info(self):
        avail_text = "Available" if self.available else "Not available"
        print(f"Title : {self.title}")
        print(f"Author: {self.author}")
        print(f"ISBN  : {self.get_isbn()}")
        print(f"Status: {avail_text}")
print("-" * 30)


class Member:
    def _init_(self, name: str, membership_id: str):
        self.name = name
        self.__membership_id = membership_id   # private attribute
        self.borrowed_books: List[Book] = []

    # getter for membership_id
    def get_membership_id(self) -> str:
        return self.__membership_id

    # setter for membership_id
    def set_membership_id(self, new_id: str):
        self.__membership_id = new_id

    def borrow_book(self, book: Book) -> bool:
        """
        Attempts to borrow a book.
        Returns True if successful, False otherwise.
        """
        if book.available:
            book.available = False
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'.")
            return True
        else:
            print(f"Cannot borrow '{book.title}': not available.")
            return False

    def return_book(self, book: Book) -> bool:
Returns a borrowed book.
        Returns True if successful, False if the member didn't have it.
        """
        if book in self.borrowed_books:
            book.available = True
            self.borrowed_books.remove(book)
            print(f"{self.name} returned '{book.title}'.")
            return True
        else:
            print(f"{self.name} has not borrowed '{book.title}'.")
            return False

    def list_borrowed(self):
        if not self.borrowed_books:
            print(f"{self.name} has no borrowed books.")
        else:
            print(f"{self.name} has borrowed:")
            for b in self.borrowed_books:
                print(f" - {b.title} (ISBN: {b.get_isbn()})")


class StaffMember(Member):
    def _init_(self, name: str, membership_id: str, staff_id: str):
        super()._init_(name, membership_id)
        self.staff_id = staff_id

    def add_book(self, library: "Library", book: Book):
        library.add_book(book)
        print(f"Staff {self.name} added book '{book.title}' to library.")


class Library:
    def _init_(self, name: str):
        self.name = name
        self.books: List[Book] = []

    def add_book(self, book: Book):
        # Prevent adding duplicates by ISBN
        if self.find_book_by_isbn(book.get_isbn()) is None:
            self.books.append(book)
        else:
            print(f"Book with ISBN {book.get_isbn()} already exists in library.")

    def find_book_by_isbn(self, isbn: str):
        for b in self.books:
            if b.get_isbn() == isbn:
                return b
        return None

    def list_books(self):
        if not self.books:
            print("Library has no books.")
            return
        print(f"Books in {self.name}:")
        for b in self.books:
            b.display_info()


# --------- Example usage ----------
if _name_ == "_main_":
    lib = Library("Central Library")

# create books
    b1 = Book("1984", "George Orwell", "978-0451524935")
    b2 = Book("To Kill a Mockingbird", "Harper Lee", "978-0061120084")
    b3 = Book("Clean Code", "Robert C. Martin", "978-0132350884")

    # add books to library (via staff)
    staff = StaffMember("Amina", "S100", "STF-001")
    staff.add_book(lib, b1)
    staff.add_book(lib, b2)
    staff.add_book(lib, b3)

    # show all books
    lib.list_books()

    # create a member and borrow/return flow
    member = Member("Khaled", "M200")
    member.borrow_book(b1)   # should succeed
    member.borrow_book(b1)   # should fail (already borrowed)
    member.list_borrowed()

    # try to find by isbn and borrow
    found = lib.find_book_by_isbn("978-0132350884")
    if found:
        member.borrow_book(found)

    # return a book
    member.return_book(b1)
    member.list_borrowed()

    # show library status after operations
    lib.list_books()
```
